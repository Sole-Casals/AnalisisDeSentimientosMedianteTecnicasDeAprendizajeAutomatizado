# Extracción de Datos Google API

```python
import googlemaps
import pandas as pd
```

```python
#Cargamos todos los municipios de España para realizar la búsqueda de valoraciones
cities_df = pd.read_excel("municipios_export_20221002_142042.xlsx")
cities_df
```

```python
#Inicialización API
gmaps = googlemaps.Client(key='***********')
```

```python
#Función para devolver las reviews de restaurantes por Municipio 
def getReviews(ca):
    CA_df = cities_df[cities_df['COMUNIDAD_AUTONOMA'].str.fullmatch(ca)]
    cities = CA_df['DENOMINACION'].to_numpy()
    cities = list(cities)
    reviews = []
    for city in cities:
        print("Leyendo Datos de %s"%(city))
        place_name = 'restaurantes cerca de %s'%(city)
        place_details = gmaps.places(place_name)
        for i in range(len(place_details['results'])):
            placeId = place_details['results'][i]['place_id']
            place = gmaps.place(placeId)
            #print(len(place['result']['reviews']))
            try:
                revs = place['result']['reviews']
                for review in place['result']['reviews']:
                    reviews.append(review)
            except:
                local = place['result']['name']
                e = 'Local "%s" del municipio "%s" no tiene ninguna valoración.'%(local,city)
                print(e)
    reviews_df =  pd.DataFrame(reviews)
    reviews_df.to_csv('%s.csv'%(ca))
```
