```python
import pandas as pd
import re
pd.set_option('display.max_colwidth', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
reviews_df = pd.read_csv("reviewsFinal_Actualizado.csv")
reviews_df = reviews_df.drop('Unnamed: 0', axis=1)
reviews_df
```

```python
reviews_df = reviews_df.drop_duplicates()
```

```python
#Comprobamos si existen menciones "@" dentro de los comentarios
reviews_df[reviews_df["Comentario"].str.contains("@[A-Za-z0-9]+\s?")].head(3)
```

```python
#Comprobamos si existen saltos de linea "\n" dentro de los comentarios
reviews_df[reviews_df["Comentario"].str.contains("\n")].head(3)
```

```python
#Comprobamos si existen hashtags "#" dentro de los comentarios
reviews_df[reviews_df["Comentario"].str.contains("#[A-Za-z0-9]+\s?")].head(3)
```

```python
#Comprobamos si existen urls "http..." dentro de los comentarios
reviews_df[reviews_df["Comentario"].str.contains("https?://[^ ]+\s?")].head(3)
```

```python
#Comprobamos si existen signos de puntuacion/ caracteres especiales dentro de los comentarios
reviews_df[reviews_df["Comentario"].str.contains("[^\w\s]")].head(3)
```

```python
#Comprobamos si existen caracteres repetidos dentro de los comentarios
reviews_df[reviews_df["Comentario"].str.contains("(.)\\1{2,}")].head(3)
```

```python
#Comprobamos si existen palabras compuestas mediante guion repetidos dentro de los comentarios
reviews_df[reviews_df["Comentario"].str.contains("([A-Za-z]+)-([A-Za-z]+)")].head(3)
```

```python
#Implementamos una funcion para poder tratar estas casuisticas y transformarlas a texto simplificado
def preprosess(text):
    text = text.lower()
    text = re.sub('@[A-Za-z0-9]+\s?','',text) #menciones
    text = re.sub('\n','',text) #saltos de linea
    text = re.sub('#[A-Za-z0-9]+\s?','',text) #hashtags
    text = re.sub(',','',text) #comas
    text = re.sub('[^\w\s]','',text) #puntuacion
    text = re.sub('"(.)\\1{2,}"','\\1\\1',text) #caracteres repetidos
    text = re.sub("([A-Za-z]+)-([A-Za-z]+)", "\\1\\2", text) #palabras compuestas
    return text
```

```python
reviews_df["Comentario"] = reviews_df["Comentario"].apply(preprosess)
reviews_df
```

```python
#Lista de los comentarios de la columna Comentario
comentarios = reviews_df["Comentario"].values.tolist()
comentarios
```

```python
#Tokenizamos los comentarios obtenidos del dataset
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

comentarios_token = []
for com in comentarios:
    comentarios_token.append(word_tokenize(com))

comentarios_token[0]
```

```python
#Utilizamos la libreria SymSpell para definir una funcion que corrije gramaticalmente los comentarios tokenizadas
from symspellpy import SymSpell, Verbosity
import pkg_resources
# load a dictionary (this one consists of 82,765 English words)
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)
dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)
# term_index: column of the term 
# count_index: column of the term's frequency
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
def CorrectText(input_term):
  # look up suggestions for multi-word input strings 
  suggestions = sym_spell.lookup_compound( 
      phrase=input_term,  
      max_edit_distance=2,  
      transfer_casing=True,  
      ignore_term_with_digits=True, 
      ignore_non_words=True, 
      split_by_space=True 
  )
  # display the correction
  for suggestion in suggestions: 
      return suggestion.term
```

```python
#Llamamos a la funcion correctora
comentarios_correct = []
for com in comentarios_token:
    l = []
    for word in com:
        l.append(CorrectText(word))
    
    comentarios_correct.append(l)
comentarios_correct
```

```python
#Eliminamos las stopwords existentes
from nltk.corpus import stopwords
nltk.download('stopwords')


stop = set(stopwords.words('english'))
# Inicializa la lista
comentarios_stop = []
# Solo guarda en la lista las palabras que no son stopwords
for com in comentarios_correct:
 l = [c for c in com if not c in stop]
 comentarios_stop.append(l)

```

```python
comentariosStopString = []
for arrStop in comentarios_stop:
    s = " ".join(arrStop)
    comentariosStopString.append(s)
    
comentariosStopString
```

```python
#Modulo PorterStemmer para obtener la raiz de las palabras
from nltk.stem import PorterStemmer

ps = PorterStemmer()
comentarios_stem = []

for com in comentarios_stop:
 l = [ps.stem(c) for c in com]
 comentarios_stem.append(l)
```

```python
comentarios_clean = []
for com in comentarios_stem:
    comentarios_clean.append(" ".join(com))
    
comentarios_clean
```

```python
#reviews_df.assign(Preprocedado = comentarios_clean)
reviews_df["Texto Preprocesado"] = comentarios_clean
reviews_df["StopWords"] = comentariosStopString
reviews_df
```

```python
reviews_df.to_csv("reviews_preproc.csv")
```
